```python
#Step 3. Assign it to a variable called online_rt
online_rt = pd.read_csv("C:\\Users\\91982\\Desktop\\Data Science, Analysis\\Assignments\\online_retail\\Online_Retail.csv", encoding="latin1")
online_rt.sample(10)
```

```python
online_rt.columns = [col.lower() for col in online_rt.columns]
```

```python
#Step 4. Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK
ex_uk = online_rt[online_rt["country"] != "United Kingdom"]
top_ord_count = ex_uk.groupby("country")["quantity"].sum().sort_values(ascending=False).head(10)
top_ord_count.plot(kind="bar", color="darkblue")
plt.xlabel("Country")
plt.ylabel("Quantity")
plt.title("Most Orders Country Wise Except UK")
```

```python
#Step 5. Exclude negative Quantity entries
online_rt = online_rt[online_rt["quantity"] > 0]
ex_uk_pos = ex_uk[ex_uk["quantity"] > 0]
```

```python
#Step 6. Create a scatterplot with the Quantity per UnitPrice by CustomerID for the top 3 Countries (except UK)
top_3_count = ex_uk_pos[ex_uk_pos["country"].isin(["Netherlands", "EIRE", "Germany"])]
top_3_count["ratio_q_up"] = top_3_count["quantity"]/top_3_count["unitprice"]
group = top_3_count.groupby("customerid")["ratio_q_up"].mean().reset_index(name="mean_ratio")
plt.figure(figsize=(15,6))
plt.scatter(group["customerid"], group["mean_ratio"], color="maroon")
plt.xlabel("Customer Id")
plt.ylabel("Quantity per Unit Priice")
plt.title("Scatterplot of Quantity per UnitPrice by CustomerID")
```

```python
#Step 7.1 Look at the first line of code in Step 6. And try to figure out if it leads to any kind of problem.
#Step 7.1.1 Display the first few rows of that DataFrame.
top_3_count.head()
```

```python
#Step 7.1.2 Think about what that piece of code does and display the dtype of UnitPrice
top_3_count.ratio_q_up.dtype
```

```python
#Step 7.1.3 Pull data from online_rt for CustomerID s 12346.0 and 12347.0.
online_rt[online_rt["customerid"].isin([12346.0, 12347.0])]
```

```python
#Step 7.3.1 Add a column to online_rt called Revenue calculate the revenue (Quantity * UnitPrice) from each sale.
online_rt["revenue"] = online_rt["quantity"] * online_rt["unitprice"]
online_rt
```

```python
#Step 7.3.2 Group by CustomerID and Country and find out the average price( AvgPrice ) each customer spends per unit.
avg_price = online_rt.groupby(["customerid", "country"])["revenue"].mean()
avg_price = avg_price.reset_index(name="avg_spend_per_unit")
avg_price
```

```python
plt.figure(figsize=(20,6))
plt.scatter(avg_price["customerid"], avg_price["avg_spend_per_unit"])
```

```python
#8.1 Group UnitPrice by intervals of 1 for prices [0,50), and sum Quantity and Revenue .
intervals = pd.interval_range(start=0, end=50, freq=1, closed="left")
online_rt["price_interval"] = pd.cut(online_rt["unitprice"], intervals)
online_rt
```

```python
top_3_count = online_rt[online_rt["country"].isin(["Germany", "Netherlands", "EIRE"])]
top_3_count = top_3_count.groupby("customerid")["revenue"].mean().reset_index(name="avg_rev")
top_3_count["price_interval"] = pd.cut(top_3_count["avg_rev"], intervals)
top_3_count.dropna(inplace=True)
top_3_count
```

```python
plt.figure(figsize=(10,6))
plt.scatter(top_3_count["price_interval"].astype(str), top_3_count["avg_rev"])
plt.xlabel("Price Interval")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
```
